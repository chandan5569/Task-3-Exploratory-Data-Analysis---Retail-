```python
# importing the libraries

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
data = pd.read_csv("SampleSuperstore.csv")
```

```python
data
```

```python
data.head()
```

# Data cleaning

```python
#Exploratory Data Analysis

#Dealing with the Missing Data in the dataset

#DATA CLEANING PART

sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#checking in the Category feild.

data.groupby(["Category"]).mean()
```

```python
data.groupby(["Category"]).max()
```

```python
data.groupby(["Category"]).min()
```

```python
#Finding the totalsum and visualising the total profit as compare to the sells.


plt.figure(figsize= (10,20))

data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()

plt.show()

#Profit for the Furniture selling price is very less.
```

```python
#Checking in the sub category feild

data.groupby('Sub-Category').mean()

#what we can clearly observe that the while selling Bookcases, Supplies, Tables is in negetive profit and these falls under the furniture category.
```

```python
data.groupby('Sub-Category').max()
```

```python
data.groupby('Sub-Category').min()
```

```python
#for the sub category we are checking.

plt.figure(figsize= (10,25))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

### Corelation

```python

corr=data.corr()

corr

```

```python
#HEAT MAP


plt.figure(figsize=(5,6))

sns.heatmap(corr,annot=True,cmap='hot')
```

```python
#Ploting based upon the REGION and the category based.

plt.figure(figsize=(20,6))

plt.subplot(2,1,1)

sns.countplot('Region',hue='Category',data=data)

plt.title('REGION-CATEGORY')

plt.ylabel('COUNT',fontsize=12)

plt.xlabel('REGION',fontsize=12)

plt.grid(alpha=0.3)
plt.show()
```

```python
#Ploting based upon the REGION and the Sub-category based.

plt.figure(figsize=(20,6))

plt.subplot(2,1,2)

sns.countplot('Region',hue='Sub-Category',data=data)

plt.title('REGIONS-SUB CATEGORY')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
#Now the Analysis on the basis of shiping mode

sns.countplot(x=data['Ship Mode'])

#Standard class has the highest number of shipping mode.
```

```python
plt.figure(figsize=(16,12))

data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")

plt.show()

#Binders having the highest value count(15.2%)
```

```python

#Sales per Ship mode

data.groupby('Ship Mode')['Sales','Profit'].agg(['sum']).plot.bar()

plt.title('Year wise Total Sales & % of profit gained')
```

```python
#Checking fo the state wise number of delivery.

plt.figure(figsize=(13,8))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.show()

#California is the city with highest number of delivery.
```

```python
top_state=data.groupby('State').sum().sort_values('Profit',ascending=False)
top_state.head(15)
```

```python
#Pairplot

figsize=(15,10)
sns.pairplot(data,hue='Sub-Category')

#More the sales more the profit.
```
